# Data Analysis Homework 1


This script is designed to do some analysis on the words used in the body of tweets with the word "Trump" in them. First I am creating a basic cleaning function for this list which can be seen below.

```python
def quickclean(word):
    charcount = 0
    for letter in word:
        if '\\' == letter:
            charcount += 1
        elif '@' == letter:
            charcount+= 1
        elif '/' == letter:
            charcount+= 1
        elif '&' == letter:
            charcount+= 1
        
    if charcount >= 1:
        return False
    else:
        return True
```

My idea with this function is a very basic cleaning of the data, based on the premise that any "words" that are created by splitting the text file which contain these characters are not part of the body of the tweet. Therefore, they should not be appended to the clean data list which I will run the positive, negative, and stopping word count on later.

Next I imported the file and began to strip punctuation and split it into a list.

```python
f = open('Trump_Raw_Tweets_Fall22.txt', 'r')
fread = f.read()
lower_f = fread.lower().replace('!', '').replace('.', '').replace('?', '').replace(',', '').replace(';', '').split()
clean_lower = []
for word in lower_f:
    if quickclean(word) == True:
        clean_lower.append(word)
    else:
        pass
f.close()
print (clean_lower)
```

You can see the result of this basic cleaning process above, we have a list containing many strings which are all lower case words. Now we can check how many of these words are on each of the lists of positive, negative, and stopper words respectively. I got the negative word list from mkulakowski2 on github, and the stopper words from larsyencken on github.

```python
with open('positive.txt') as p:
    p_words = p.read().splitlines()
    p_words.remove('trump')
    
with open('negative.txt') as n:
    n_words = n.read().splitlines()
    
with open('stopwords.txt') as s:
    s_words = s.read().splitlines()
```

This cell imports the files and makes them into lists with each word as its own entry. Below I create the variables I will need in order to do the analysis on my cleaned data, and then counted each occurence of a positive, negative, and stop word with this simple for loop. 

I also added a line of code in after a fellow student pointed out to me that 'trump' is in the list of positive words! That means without removing it, the script is counting 'trump' as a positive word while in this data-set it should be considered an other word or just removed completely. Not doing this has significant impact on the analysis.

```python
posnum = 0
negnum = 0
stopnum = 0
total_words = 0
other_words = 0
for word in clean_lower:
    total_words += 1
    if word in p_words:
        posnum += 1
    elif word in n_words:
        negnum += 1
    elif word in s_words:
        stopnum += 1
    else:
        other_words += 1
```

Now in the cell below we print all of our results:

```python
print("Total positive words: %s" % posnum)
print("Total negative words: %s" % negnum)
print("Total stop words: %s" % stopnum)
print('total other words: %s' % other_words)

pos_percent = "{:.0%}".format(posnum/total_words)
neg_percent = "{:.0%}".format(negnum/total_words)
stop_percent = "{:.0%}".format(stopnum/total_words)
other_percent = "{:.0%}".format(other_words/total_words)
ratio = posnum/negnum

print('Positive words make up %s of the total words' % pos_percent)
print('Negative words make up %s of the total words' % neg_percent)
print('Stop words make up %s of the total words' % stop_percent)
print('Other words make up %s of the total words' % other_percent)

print('The ratio of positive to negative words is: %s' % ratio)
```

As is seen above there are more negative words than positive words by a significant factor of nearly 100%. It should be noted that if you do not remove the word 'trump' from the positive words dataset, you will get results that put positive ahead of negative. From this analysis I would conclude that sentiment is generally more negative than positive surrounding Trump on Twitter, which matches up with my experience on Twitter.
